```python
import numpy as np
import torch
import torch.nn as nn
# from torch.utils.tensorboard import SummaryWriter

import matplotlib.pyplot as plt
from scipy import interpolate
import matplotlib
import scienceplots

batch_loss_curve = np.load(file='b_checkpoints/ChangeFormerV5/batch_train_loss.npy')
# epoch_loss_curve = np.load(file='./ChangeFormerV5/epoch_train_loss.npy')

# print(loss_curve)

# def tensor_board():
#     batch_writer = SummaryWriter('Batch Loss Curve')
#     epoch_writer = SummaryWriter('Epoch Loss Curve')

#     for i, loss in enumerate(batch_loss_curve):
#         # print(i, loss)
#         batch_writer.add_scalar(tag='batch loss', scalar_value=loss, global_step=i)

#     for i, loss in enumerate(epoch_loss_curve):
#         print(i, loss)
#         epoch_writer.add_scalar(tag='epoch loss', scalar_value=loss, global_step=i)


#     batch_writer.close()

# def draw_plot():
#     data = np.random.random(1000)
#     print(data.shape)
#     plt.plot(batch_loss_curve)
#     plt.xlabel('x label')
#     plt.ylabel('y label')
#     plt.show()
#     plt.title('示例曲线图')

def ieee_plot():
    
    x=[0,1,5,10,15,20,30,40,50,60,65,70]
    y1=[-3.26,-3.07,-2.28,-1.27,-0.33,0.47,1.64,2.36,2.67,2.73,2.72,2.68]
    y2=[-3.256,-3.064,-2.283,-1.273,-0.332,0.468,1.638,2.363,2.675,2.735,2.716,2.681]
    y3=np.cos(np.linspace(0, 4 * np.pi, len(x)))

    # 样条插值y1,返回参数tck,返回参数u
    tck,u = interpolate.splprep([x,y1], s=0)    # splprep传入x和y坐标，生成插值函数，s表示平滑度，0为完全通过输入数据点，tck和u为返回参数
    unew = np.arange(0, 1.01, 0.01)             # 修改参数 u -> unew 在单位1区间以更密集的形式进行插值
    out = interpolate.splev(unew, tck)          # 插值splev(x, tck, der=0, ext=0)
    # 样条插值y2
    tck,u = interpolate.splprep([x,y2], s=0)
    unew = np.arange(0, 1.01, 0.01)
    out1 = interpolate.splev(unew, tck)
    # 样条插值y2
    tck, u = interpolate.splprep([x, y3], s=0)
    unew = np.arange(0, 1.01, 0.01)
    out2 = interpolate.splev(unew, tck)

    # 生成自己的数据点
    x_axis = np.linspace(start=0, stop=len(batch_loss_curve), num=len(batch_loss_curve))
    random_loss_curve = np.random.random(len(batch_loss_curve))
    print(x_axis.shape)
    print(random_loss_curve.shape)
    tck, u = interpolate.splprep([x_axis,random_loss_curve], s=0)
    unew = np.linspace(start=0, stop=1, num=10)
    x_axis_new, my_out = interpolate.splev(unew, tck)

    # print(plt.style.available)

    # 绘制
    if True:
        """
        plt.subplots()创建子图  plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
        返回整个整个图标对象fig参数和每一个子图对象ax参数
        ax.plot进行画图
        ax.set_title对每个图表设置标题
        fig.subtitle对整个图标进行设置
        plt.tight_layout()调整子图之间的间距，以确保内容不会重叠
        ax.legend()对图例曲线加入标识
        """
        # with plt.style.context(['seaborn-paper','seaborn-dark']):
        with plt.style.context(['classic','ieee','no-latex']):
            fig, ax = plt.subplots()                       
            ax.plot(out[0], out[1], label='smartload')        
            ax.plot(out1[0], out1[1], label='lr')     
            ax.plot(out2[0], out2[1], label='cosine')
            ax.legend(title='Order', prop={'size': 5})        
            ax.set(xlabel='Heel angle (°)')        
            ax.set(ylabel='Gz (m)')        
            ax.autoscale(tight=True) #设置Y刻度最大值
            ax.set_ylim(top=3)          
            # fig.savefig('b_checkpoints/ChangeFormerV5/fig1.jpg', dpi=300)

            plt.show()

    if False:
        with plt.style.context(['science', 'ieee']):
            plt.plot(x_axis_new, my_out, '-', label='插值')
            plt.plot(x_axis, random_loss_curve, 'o', label='原始')
            plt.xlabel('epoches')
            plt.ylabel('loss')
            # plt.axis('test')
            plt.show()



# draw_plot()
if __name__ == "__main__":
    # draw_plot()
    ieee_plot()
    # print(matplotlib.get_configdir())

```
